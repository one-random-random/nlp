# Stop Words
In this practical we will explore removing stop words using the NLTK and SpaCy libraries.

## NLTK

```python
import nltk
nltk.download('punkt_tab')

from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

text = "When the setting sun hits Horsetail Fall, it creates an optical illusion that transforms the water coming off the rocks into what looks like molten lava pouring down the granite face of El Capitan. According to the US National Park Service, the fiery effect appears only in mid to late February, and only on clear evenings when the waterfall is flowing. Horsetail Fall, in California's Yosemite National Park, typically runs only during winter, and even slight haze or cloud cover can greatly diminish the display, according to the NPS."
text_tokens = word_tokenize(text)

tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]

print(tokens_without_sw)
```

In this example you can see that 'to', 'in', 'he', 'is', 'not', 'too', 'of' have been removed from the text. These tokens can also be joined to output a full sentence:

```python
filtered_sentence = (" ").join(tokens_without_sw)
print(filtered_sentence)
```

It is possible to add words to the 'stop words' list, you can also print the current list of stop words.

```python
print(len(stopwords.words('english')))
print(stopwords.words('english'))
```

In the exampe provided, if we wanted to add the word 'write' as a stop word, we can do this using the following code:

```python
all_stopwords = stopwords.words('english')
all_stopwords.append('write')

text_tokens = word_tokenize(text)
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords]

print(tokens_without_sw)
```

We can also add a list of stop words using the following code:

```python
sw_list = ['hits','mid', 'late', 'runs']
all_stopwords.extend(sw_list)

text_tokens = word_tokenize(text)
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords]

print(tokens_without_sw)
```

In this case, removing the word 'not' changes the meaning of the sentence. If we want to remove a stop word, this can be completed using the following code:

```python
all_stopwords = stopwords.words('english')
all_stopwords.remove('not')

text_tokens = word_tokenize(text)
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords]

print(tokens_without_sw)
```

## spaCy
Now we will try completing the same tasks using SpaCy.

```python
# Import spaCy and load the English language library
import spacy
# This will take a while to load initially
sp = spacy.load("en_core_web_sm")
all_stopwords = sp.Defaults.stop_words

text = "When the setting sun hits Horsetail Fall, it creates an optical illusion that transforms the water coming off the rocks into what looks like molten lava pouring down the granite face of El Capitan. According to the US National Park Service, the fiery effect appears only in mid to late February, and only on clear evenings when the waterfall is flowing. Horsetail Fall, in California's Yosemite National Park, typically runs only during winter, and even slight haze or cloud cover can greatly diminish the display, according to the NPS."
text_tokens = word_tokenize(text)
tokens_without_sw= [word for word in text_tokens if not word in all_stopwords]

print(tokens_without_sw)
```

In SpaCy there are more stop words defined, 326 stop words compared to the 179 in NLTK.

```python
print(len(all_stopwords))
print(all_stopwords)
```

It is also possible to add stop words either individually or through using an array, see below for an individual example:

```python
all_stopwords = sp.Defaults.stop_words
all_stopwords.add("Horsetail")

text = "When the setting sun hits Horsetail Fall, it creates an optical illusion that transforms the water coming off the rocks into what looks like molten lava pouring down the granite face of El Capitan. According to the US National Park Service, the fiery effect appears only in mid to late February, and only on clear evenings when the waterfall is flowing. Horsetail Fall, in California's Yosemite National Park, typically runs only during winter, and even slight haze or cloud cover can greatly diminish the display, according to the NPS."
text_tokens = word_tokenize(text)
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords]

print(tokens_without_sw)
```

We can also remove stop words from the list.

```python
all_stopwords = sp.Defaults.stop_words
all_stopwords.remove('to')

text = "When the setting sun hits Horsetail Fall, it creates an optical illusion that transforms the water coming off the rocks into what looks like molten lava pouring down the granite face of El Capitan. According to the US National Park Service, the fiery effect appears only in mid to late February, and only on clear evenings when the waterfall is flowing. Horsetail Fall, in California's Yosemite National Park, typically runs only during winter, and even slight haze or cloud cover can greatly diminish the display, according to the NPS."
text_tokens = word_tokenize(text)
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords]

print(tokens_without_sw)
```
